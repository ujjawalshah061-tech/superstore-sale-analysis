```python
# Converting sales column to numeric for proper mathematical analysis
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')

# Converting order_date to datetime format for time-based analysis (trends, seasonality)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)

# Converting ship_date to datetime format for delivery and shipping analysis
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce', dayfirst=True)
```

```python
df['delivery_days'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
df['profit_ratio'] = df['profit'] / df['sales'].replace(0, np.nan)
```

```python

extreme_cases = df[(df['profit'] > 4000) | (df['profit'] < -3000)]
print(extreme_cases[['customer_name', 'product_name', 'sales', 'profit', 'discount']])
```

```python
# Analysis of Extreme Profit/Loss Orders
print("Top 5 Profit-Making Orders:")
display(extreme_cases[extreme_cases['profit'] > 0].sort_values(by='profit', ascending=False))

print("\nTop 5 Loss-Making Orders:")
display(extreme_cases[extreme_cases['profit'] < 0].sort_values(by='profit'))
```
